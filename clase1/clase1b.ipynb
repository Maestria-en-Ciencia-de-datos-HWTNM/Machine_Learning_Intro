```python
# Importar pandas (por convención se usa el alias 'pd')
import pandas as pd
import numpy as np

print('Versión de Pandas:', pd.__version__)
print('Versión de NumPy:', np.__version__)
```

## 1. Pandas Series

Una **Series** es una estructura de datos unidimensional similar a una columna en una hoja de cálculo. Cada elemento tiene un índice (etiqueta) y un valor.

```python
# Crear una Series a partir de una lista
temperaturas = pd.Series([22.5, 24.1, 19.8, 27.3, 23.0],
                          index=['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes'])
temperaturas
```

```python
# Acceder a elementos por etiqueta y por posición
print('Temperatura del Lunes:', temperaturas['Lunes'])
print('Temperatura del tercer día:', temperaturas.iloc[2])
```

```python
# Operaciones estadísticas sobre una Series
print(f'Media:           {temperaturas.mean():.2f}°C')
print(f'Máxima:          {temperaturas.max():.2f}°C')
print(f'Mínima:          {temperaturas.min():.2f}°C')
print(f'Desv. estándar:  {temperaturas.std():.2f}°C')
```

```python
# Crear una Series desde un diccionario
poblacion = pd.Series({
    'Ciudad de México': 21_671_908,
    'Buenos Aires':     15_369_919,
    'Lima':             11_044_607,
    'Bogotá':           10_978_360,
    'Santiago':          6_257_516
})
poblacion
```

## 2. Pandas DataFrame

Un **DataFrame** es una estructura de datos bidimensional, similar a una tabla de base de datos o una hoja de cálculo. Está compuesto por filas y columnas, donde cada columna es una Serie.

```python
# Crear un DataFrame desde un diccionario
datos = {
    'nombre':      ['Ana García', 'Carlos López', 'María Martínez', 'Pedro Sánchez', 'Laura Torres'],
    'edad':        [28, 35, 24, 42, 31],
    'ciudad':      ['Bogotá', 'Ciudad de México', 'Buenos Aires', 'Lima', 'Santiago'],
    'salario':     [4500, 6200, 3800, 7100, 5300],
    'departamento':['Ingeniería', 'Marketing', 'Diseño', 'Finanzas', 'Ingeniería']
}

df = pd.DataFrame(datos)
df
```

### 2.1 Exploración inicial del DataFrame

Cuando recibimos un nuevo dataset, lo primero es explorar su estructura y contenido:

```python
# Primeras y últimas filas
print('=== Primeras 3 filas ===')
print(df.head(3))
print('\n=== Últimas 2 filas ===')
print(df.tail(2))
```

```python
# Información general del DataFrame
df.info()
```

```python
# Estadísticas descriptivas
df.describe()
```

```python
# Forma del DataFrame (filas, columnas)
print('Forma:', df.shape)
print('Columnas:', df.columns.tolist())
print('Tipos de datos:\n', df.dtypes)
```

### 2.2 Selección de datos

Pandas ofrece múltiples formas de seleccionar datos:

```python
# Seleccionar una columna (devuelve una Series)
df['nombre']
```

```python
# Seleccionar múltiples columnas (devuelve un DataFrame)
df[['nombre', 'salario', 'departamento']]
```

```python
# Selección por posición con iloc [fila, columna]
print('Fila 2, columna 3:', df.iloc[2, 3])
print('\nPrimeras 3 filas, primeras 2 columnas:')
df.iloc[:3, :2]
```

```python
# Selección por etiqueta con loc
df.loc[0:2, ['nombre', 'ciudad', 'salario']]
```

### 2.3 Filtrado de datos (Boolean Indexing)

Una de las operaciones más poderosas de Pandas es filtrar filas según condiciones:

```python
# Filtrar empleados con salario mayor a 5000
df[df['salario'] > 5000]
```

```python
# Filtrar con múltiples condiciones (AND con &, OR con |)
df[(df['salario'] > 4000) & (df['edad'] < 35)]
```

```python
# Filtrar por valores en una lista (isin)
df[df['departamento'].isin(['Ingeniería', 'Diseño'])]
```

### 2.4 Agregar y modificar columnas

```python
# Agregar nueva columna calculada
df['salario_anual'] = df['salario'] * 12
df[['nombre', 'salario', 'salario_anual']]
```

```python
# Aplicar una función a una columna con apply
def categoria_edad(edad):
    if edad < 30:
        return 'Junior'
    elif edad < 40:
        return 'Senior'
    else:
        return 'Experto'

df['categoria'] = df['edad'].apply(categoria_edad)
df[['nombre', 'edad', 'categoria']]
```

## 3. Manejo de Valores Faltantes (NaN)

Los datos del mundo real raramente están completos. Pandas facilita la detección y manejo de valores faltantes:

```python
# Crear DataFrame con valores faltantes
df_incompleto = pd.DataFrame({
    'producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares'],
    'precio':   [1200.0, 25.0, None, 450.0, None],
    'stock':    [10, None, 50, 8, 15],
    'categoria':['Computadoras', 'Periféricos', 'Periféricos', None, 'Audio']
})
df_incompleto
```

```python
# Detectar valores faltantes
print('¿Hay valores nulos?')
print(df_incompleto.isnull())
print('\nTotal de nulos por columna:')
print(df_incompleto.isnull().sum())
```

```python
# Rellenar valores faltantes con: 
df_relleno = df_incompleto.copy()
df_relleno['precio'] = df_relleno['precio'].fillna(df_relleno['precio'].mean()) #la media de la columna
df_relleno['stock'] = df_relleno['stock'].fillna(df_relleno['stock'].median())#la mediana de la columna
df_relleno['categoria'] = df_relleno['categoria'].fillna('Sin categoría')# la frase "Sin categoría"
df_relleno
```

```python
# Eliminar filas con valores faltantes
df_limpio = df_incompleto.dropna()
df_limpio
```

## 4. Agrupación y Agregación (groupby)

La operación `groupby` es fundamental en análisis de datos. Permite dividir los datos en grupos y aplicar funciones de agregación:

```python
# Retomamos el DataFrame de empleados
# Salario promedio por departamento
df.groupby('departamento')['salario'].mean().round(2)
```

```python
# Múltiples agregaciones a la vez
df.groupby('departamento')['salario'].agg(['mean', 'min', 'max', 'count']).round(2)
```

```python
# Agrupar por múltiples columnas
df.groupby(['departamento', 'categoria'])[['salario', 'edad']].mean().round(2)
```

## 5. Ordenamiento y Clasificación

```python
# Ordenar por salario de mayor a menor
df.sort_values('salario', ascending=False)[['nombre', 'salario', 'departamento']]
```

```python
# Ordenar por múltiples columnas
df.sort_values(['departamento', 'salario'], ascending=[True, False])[['nombre', 'departamento', 'salario']]
```

## 6. Carga de Datos desde CSV

En la práctica, los datos provienen de archivos externos. Aquí simulamos la carga de un CSV usando `pd.read_csv()`. También podemos crear y exportar DataFrames:

```python
# Simular un dataset de ventas más grande
np.random.seed(42)
n = 100

productos = ['Laptop', 'Mouse', 'Teclado', 'Monitor', 'Auriculares', 'Webcam', 'USB Hub']
regiones  = ['Norte', 'Sur', 'Este', 'Oeste', 'Centro']

df_ventas = pd.DataFrame({
    'fecha':    pd.date_range(start='2024-01-01', periods=n, freq='3D'),
    'producto': np.random.choice(productos, n),
    'region':   np.random.choice(regiones, n),
    'cantidad': np.random.randint(1, 20, n),
    'precio_unitario': np.random.choice([25, 50, 75, 100, 150, 450, 1200], n)
})

df_ventas['ingreso_total'] = df_ventas['cantidad'] * df_ventas['precio_unitario']
df_ventas.head(10)
```

```python
# Guardar el DataFrame como CSV
df_ventas.to_csv('ventas_2024.csv', index=False)
print('Archivo guardado: ventas_2024.csv')

# Cargar el CSV
df_cargado = pd.read_csv('ventas_2024.csv', parse_dates=['fecha'])
print('\nDataFrame cargado:')
df_cargado.info()
```

## 7. Análisis del Dataset de Ventas

```python
# Ingresos totales por producto
ingresos_producto = df_ventas.groupby('producto')['ingreso_total'].sum().sort_values(ascending=False)
ingresos_producto
```

```python
# Tabla pivote: ingresos por producto y región
tabla_pivote = df_ventas.pivot_table(
    values='ingreso_total',
    index='producto',
    columns='region',
    aggfunc='sum',
    fill_value=0
)
tabla_pivote
```

```python
# Tendencia mensual de ventas
df_ventas['mes'] = df_ventas['fecha'].dt.to_period('M')
ventas_mensuales = df_ventas.groupby('mes')['ingreso_total'].sum()
ventas_mensuales
```

## 8. Combinación de DataFrames (Merge y Concat)

Es común necesitar combinar información de múltiples tablas:

```python
# Crear dos DataFrames relacionados
clientes = pd.DataFrame({
    'cliente_id': [1, 2, 3, 4, 5],
    'nombre':     ['Ana', 'Luis', 'Sofía', 'Jorge', 'Elena'],
    'ciudad':     ['Bogotá', 'Lima', 'Santiago', 'Ciudad de México', 'Buenos Aires']
})

pedidos = pd.DataFrame({
    'pedido_id':  [101, 102, 103, 104, 105, 106],
    'cliente_id': [1, 3, 2, 1, 5, 4],
    'monto':      [250, 480, 120, 375, 610, 89]
})

print('Clientes:')
print(clientes)
print('\nPedidos:')
print(pedidos)
```

```python
# INNER JOIN: solo clientes con pedidos
merged = pd.merge(pedidos, clientes, on='cliente_id', how='inner')
merged
```

```python
# LEFT JOIN: todos los clientes, aunque no tengan pedidos
left_merged = pd.merge(clientes, pedidos, on='cliente_id', how='left')
left_merged
```

```python
# Concatenar DataFrames verticalmente
df_extra = pd.DataFrame({
    'nombre':      ['Roberto Díaz'],
    'edad':        [29],
    'ciudad':      ['Montevideo'],
    'salario':     [4900],
    'departamento':['Ingeniería'],
    'salario_anual':[58800],
    'categoria':   ['Junior']
})

df_total = pd.concat([df, df_extra], ignore_index=True)
df_total
```
